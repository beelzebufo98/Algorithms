```python
#A1i
import pandas as pd
import matplotlib.pyplot as plt


file_path = 'results_area.csv'
data = pd.read_csv(file_path)

exact_area =  0.944517

data['WideDeviation'] = abs(data['WideArea'] - exact_area) / exact_area * 100
data['NarrowDeviation'] = abs(data['NarrowArea'] - exact_area) / exact_area * 100

# График 1: Приближенные значения площадей
plt.figure(figsize=(12, 6))
plt.plot(data['Points'], data['WideArea'], label='Wide Area', color='blue')
plt.plot(data['Points'], data['NarrowArea'], label='Narrow Area', color='green')
plt.axhline(y=exact_area, color='red', linestyle='--', label='Exact Area (0.94557)')
plt.title('Approximate Area vs Number of Points')
plt.xlabel('Number of Points (N)')
plt.ylabel('Area')
plt.legend()
plt.grid()
plt.savefig("number1_points.png")
plt.show()

# График 2: Относительные отклонения
plt.figure(figsize=(12, 6))
plt.plot(data['Points'], data['WideDeviation'], label='Wide Area Deviation', color='blue')
plt.plot(data['Points'], data['NarrowDeviation'], label='Narrow Area Deviation', color='green')
plt.axhline(y=0, color='red', linestyle='--', label='Exact Deviation (0)')
plt.title('Relative Deviation from Exact Area vs Number of Points')
plt.xlabel('Number of Points (N)')
plt.ylabel('Relative Deviation (%)')
plt.legend()
plt.grid()
plt.savefig("number2_points.png")
plt.show()
```

```python
#A2i
import matplotlib.pyplot as plt
import pandas as pd

merge_data = pd.read_csv("merge_results.csv")
hybrid_data = pd.read_csv("hybrid_results.csv")


def plot_and_save(x, y, title, filename, xlabel="Array Size", ylabel="Time (ms)", color='blue'):
    plt.figure(figsize=(8, 5))
    plt.plot(x, y, label=title, color=color, marker='o')
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.legend()
    plt.grid()
    plt.savefig(filename, format="jpg", dpi=300)
    plt.show()
    plt.close()

# Построение и сохранение графиков для Merge Sort
plot_and_save(merge_data["Size"], merge_data["Random"], "Merge Sort - Random Array", "merge_sort_random.jpg", color="blue")
plot_and_save(merge_data["Size"], merge_data["Decrease"], "Merge Sort - Decrease Array", "merge_sort_decrease.jpg", color="green")
plot_and_save(merge_data["Size"], merge_data["NearSorted"], "Merge Sort - Near Sorted Array", "merge_sort_near_sorted.jpg", color="red")

# Построение и сохранение графиков для Hybrid Sort
plot_and_save(hybrid_data["Size"], hybrid_data["Random"], "Hybrid Sort - Random Array", "hybrid_sort_random.jpg", color="blue")
plot_and_save(hybrid_data["Size"], hybrid_data["Decrease"], "Hybrid Sort - Decrease Array", "hybrid_sort_decrease.jpg", color="green")
plot_and_save(hybrid_data["Size"], hybrid_data["NearSorted"], "Hybrid Sort - Near Sorted Array", "hybrid_sort_near_sorted.jpg", color="red")
```
